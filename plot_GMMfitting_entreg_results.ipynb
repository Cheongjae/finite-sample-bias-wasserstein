```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import pickle
```

```python
for k, seed in enumerate(seed_set):
    for j, mb in enumerate(mb_set):
        for i, dim in enumerate(dim_set):
            l = 0
            foldername = f'W_D{dim}K4_mb{mb}_ds42_reg{reg}_pifixed_kmeansinit_lr0.01_T50000_S{seed}'
            filename = './results/'+foldername+'/results.pickle'
            # load from pickle file
            with open(filename, "rb") as f:
                results[foldername] = pickle.load(f)
            score_set_v3[l,i,j,k] = results[foldername]['score'][-1]
            score_kmeans_set_v3[l,i,j,k] = results[foldername]['score_kmeans']
            score_gmm_set_v3[l,i,j,k] = results[foldername]['score_GMM']

            l = 1
            foldername = f'W_D{dim}K4_mb{mb}_ds42_debias_reg{reg}_pifixed_kmeansinit_lr0.01_T50000_S{seed}'
            filename = './results/'+foldername+'/results.pickle'
            # load from pickle file
            with open(filename, "rb") as f:
                results[foldername] = pickle.load(f)
            score_set_v3[l,i,j,k] = results[foldername]['score'][-1]
            score_kmeans_set_v3[l,i,j,k] = results[foldername]['score_kmeans']
            score_gmm_set_v3[l,i,j,k] = results[foldername]['score_GMM']

            l = 2
            foldername = f'W_D{dim}K4_mb{mb}_ds42_reg{reg}_pifixed_fixdata_kmeansinit_lr0.01_T50000_S{seed}'
            filename = './results/'+foldername+'/results.pickle'
            # load from pickle file
            with open(filename, "rb") as f:
                results[foldername] = pickle.load(f)
            score_set_v3[l,i,j,k] = results[foldername]['score'][-1]
            score_kmeans_set_v3[l,i,j,k] = results[foldername]['score_kmeans']
            score_gmm_set_v3[l,i,j,k] = results[foldername]['score_GMM']

            l = 3
            foldername = f'W_D{dim}K4_mb{mb}_ds42_debias_reg{reg}_pifixed_fixdata_kmeansinit_lr0.01_T50000_S{seed}'
            filename = './results/'+foldername+'/results.pickle'
            # load from pickle file
            with open(filename, "rb") as f:
                results[foldername] = pickle.load(f)
            score_set_v3[l,i,j,k] = results[foldername]['score'][-1]
            score_kmeans_set_v3[l,i,j,k] = results[foldername]['score_kmeans']
            score_gmm_set_v3[l,i,j,k] = results[foldername]['score_GMM']
```
